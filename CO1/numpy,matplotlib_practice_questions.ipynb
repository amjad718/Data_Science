```python
# 1.	Write a Python program that creates a NumPy array from a list of numbers and calculates the mean, median, and standard deviation of the elements.

import numpy as np

array = np.array([12, 34, 56, 78, 90, 23, 45, 67, 89])

print("NumPy Array:", array)
print("Mean:", np.mean(array))
print("Median:", np.median(array))
print("Standard Deviation:", np.std(array))
```

```python
# 2. Find the number of occurrences of a sequence in a NumPy array
import numpy as np

arr = np.array([[2,8,9,4],
       [9,4,9,4],
       [4,5,9,7],
       [2,9,4,3]])

sequence_to_find = np.array([9,4])

indices = np.where(np.all(np.vstack([arr[i:i+len(sequence_to_find)] == sequence_to_find for i in range(len(arr) - len(sequence_to_find) + 1)]), axis=0))
num_occurrences = len(indices[0])

# Print the result
print(f"Sequence {sequence_to_find} occurs {num_occurrences} times in the array.")
```

```python
# 3.	Find the k smallest values of a NumPy array

import numpy as np

arr = np.array([12, 34, 1, 5, 23, 9, 7, 18, 2, 45])

k = int(input("Enter the value for k"))
sorted_arr = np.sort(arr)
k_smallest = sorted_arr[:k]

# Print the k smallest values
print(f"The {k} smallest values are: {k_smallest}")
```

```python
# 4.	Calculate the sum of the diagonal elements of a NumPy array

import numpy as np

arr = np.array([[55,25,15],
                [30,44,2],
                [11,45,77]])

diagonal_sum = np.trace(arr)

print("Sum of diagonal elements:", diagonal_sum)
```

```python
# 5.	Find indices of elements equal to zero in a NumPy array

import numpy as np

# Create a NumPy array
arr = np.array([[1, 0, 3],
                [0, 5, 0],
                [7, 0, 9]])

# Find indices of elements equal to zero
zero_indices = np.argwhere(arr == 0)
print(zero_indices)
```

```python
# 6.	Calculate the Euclidean distance using NumPy

import numpy as np

point1 = np.array([1, 2, 3])
point2 = np.array([4, 5, 6])

euclidean_distance = np.linalg.norm(point1 - point2)

print("Euclidean distance between point1 and point2:", euclidean_distance)
```

```python
# 8.	Calculate the determinant and inverse of a square matrix using NumPy functions.

import numpy as np

matrix = np.array([[2,1,3],
                  [1,2,4],
                  [3,4,9]])

determinant = np.linalg.det(matrix)

# A nonzero determinant means an inverse exists
if determinant != 0:
    inverse = np.linalg.inv(matrix)
    print("Matrix Determinant:", determinant)
    print("\nMatrix Inverse:")
    print(inverse)
else:
    print("The matrix is singular, and its inverse does not exist.")
```

```python
# 10.	Simulate rolling a fair six-sided die 1000 times and count the occurrences of each number.

import numpy as np
import random
from collections import Counter

num_rolls = 1000

rolls = [random.randint(1, 6) for _ in range(num_rolls)]

occurrences = Counter(rolls)

print("Number of occurrences of each number:")
for number, count in occurrences.items():
    print(f"Number {number}: {count} times")
```

```python
# 12.	Create a scatter plot with labeled data points, adjusting the colors and markers.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
labels = ['A', 'B', 'C', 'D', 'E']

colors = ['red', 'blue', 'green', 'orange', 'purple']
markers = ['o', 's', '^', 'D', 'v']

# Create a scatter plot with labeled data points
for i in range(len(x)):
    plt.scatter(x[i], y[i], label=labels[i], color=colors[i], marker=markers[i])

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Labeled Data Points')
plt.legend()

plt.show()
```

```python
# 14.	Load an image as a NumPy array and apply a grayscale filter to it.

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image = Image.open('/content/download.jpeg')

# Convert the image to grayscale
gray_image = image.convert('L')

# Convert the grayscale image to a NumPy array
gray_array = np.array(gray_image)

# Display the original and grayscale images
plt.subplot(1, 2, 1)
plt.imshow(np.array(image))
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(gray_array, cmap='gray')
plt.title('Grayscale Image')

plt.show()
```

```python
# 15.	Load an image as a Mumpy array. Perform image manipulations such as cropping,resizing and color channel conversions. Visualize the original and modified images using Matplotlib.

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the image as a NumPy array
image = Image.open('/content/download.jpeg')  # Replace 'your_image_path.jpg' with the path to your image
original_image = np.array(image)

# Perform image manipulations
# Crop the image
crop_box = (100, 100, 400, 400)  # (left, upper, right, lower)
cropped_image = original_image[crop_box[1]:crop_box[3], crop_box[0]:crop_box[2]]

# Resize the image
new_size = (200, 200)
resized_image = image.resize(new_size)
resized_array = np.array(resized_image)

# Convert to grayscale
gray_image = image.convert('L')
gray_array = np.array(gray_image)

# Display the original and modified images
plt.figure(figsize=(12, 4))

plt.subplot(1, 4, 1)
plt.imshow(original_image)
plt.title('Original Image')

plt.subplot(1, 4, 2)
plt.imshow(cropped_image)
plt.title('Cropped Image')

plt.subplot(1, 4, 3)
plt.imshow(resized_array)
plt.title('Resized Image')

plt.subplot(1, 4, 4)
plt.imshow(gray_array, cmap='gray')
plt.title('Grayscale Image')

plt.tight_layout()
plt.show()
```

```python
# 16.	Plot a complex number in Python using Matplotlib

import matplotlib.pyplot as plt

# Create a complex number
complex_number = 1.5 + 2.5j  # 3 + 4i

# Extract the real and imaginary parts
real_part = complex_number.real
imaginary_part = complex_number.imag

# Create a scatter plot to represent the complex number
plt.figure(figsize=(6, 6))
plt.scatter(real_part, imaginary_part, color='red', marker='o', label='Complex Number')

# Set axis labels
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')

# Set axis limits for better visualization
plt.xlim(0, 5)
plt.ylim(0, 5)

# Add a legend
# plt.legend()

# Add a grid for reference
plt.grid(True)

# Add a title
plt.title('Complex Number Plot')

# Show the plot
plt.show()
```

```python
# 17.	Normal Distribution Plot using Numpy and Matplotlib

import numpy as np
import matplotlib.pyplot as plt

mean = 0
std_dev = 1
num_samples = 1000

data = np.random.normal(mean, std_dev, num_samples)

plt.hist(data, bins=30, density=True, alpha=0.6, color='b')

plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')

x = np.linspace(-3, 3, 1000)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
plt.plot(x, pdf, 'r', linewidth=2)

plt.show()
```
