```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Drawing a simple line in graph

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Plotting without line

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
# Drawing a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10)

xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
# If we do not give x-axis any values, it will take the default values as 1,2,3...

ypoints = np.array([2,5,3,9,1,6,7,4,2,6,3,4,6,7,2,4])

plt.plot(ypoints)
plt.show()
```

```python
# You can use marker to get the point in the graph

ypoints = np.array([2,5,3,9,1,6,7,4,2,6,3,4,6,7,2,4])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
ypoints = np.array([2,5,3,9,1,6,7,4,2,6,3,4,6,7,2,4])

plt.plot(ypoints, 'o-.k')
# Here we use a format instead of the markers
# first letter is the marker, second is the type of line, and the third one is the colour
plt.show()
```

```python
# We can use ms for marker size

ypoints = np.array([2,5,3,9,1,6,7,4,2,6,3,4,6,7,2,4])

plt.plot(ypoints, marker = 'o', ms=20 , mfc='r', mec='g')
plt.show()
```

```python
# Adding grid lines to a plot

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.grid(axis = 'y')

plt.show()
```

```python
# We can plot two graphs together by using subplot
#Syntax : subplot(num1,num2,num3) -> No of rows, No of colums, The order

# 1st Plot
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.subplot(2,1,1)
plt.plot(x, y)


# 2nd Plot

ypoints = np.array([2,5,3,9,1,6,7,4,2,6,3,4,6,7,2,4])

plt.subplot(2,1,2)
plt.plot(ypoints)


plt.show()
```

```python
#Drawing 6 plots all at once

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

```python
# Scatter Plot

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
# We can draw bar graphs using bar()

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
# Drawing a histogram using hist()

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

```python
# We can plot pie chart using pie()

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```
